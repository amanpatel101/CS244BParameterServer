```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fnone, f40, f50, f60, f100 = [], [], [], [], []
for it in [1, 2, 3]:
    fnone.append(json.load(open("../results/fault_tolerance/failure_no_failure/fault_tolerant_failure_nofail_try_%i.json"%(it)))["accuracy"])
    f40.append(json.load(open("../results/fault_tolerance/failure_no_failure/fault_tolerant_failure_40_try_%i.json"%(it)))["accuracy"])
    f50.append(json.load(open("../results/fault_tolerance/failure_no_failure/fault_tolerant_failure_50_try_%i.json"%(it)))["accuracy"])
    f60.append(json.load(open("../results/fault_tolerance/failure_no_failure/fault_tolerant_failure_60_try_%i.json"%(it)))["accuracy"])
    f100.append(json.load(open("../results/fault_tolerance/failure_no_failure/fault_tolerant_failure_100_try_%i.json"%(it)))["accuracy"])
```

```python
plt.figure(dpi=300)
plt.plot(np.mean(fnone, axis=0)[:100])
plt.xlabel("Iteration", size=14)
plt.ylabel("Accuracy (%)", size=14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.title("Training Curve for First 100 Iterations -\n No Failure", size=16)
plt.ylim(45, 85, 5)
plt.show()
plt.figure(dpi=300)
plt.plot(np.mean(f40, axis=0)[:100])
plt.xlabel("Iteration", size=14)
plt.ylabel("Accuracy (%)", size=14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.title("Training Curve for First 100 Iterations -\n Failure at 40", size=16)
plt.ylim(45, 85, 5)
plt.show()
plt.figure(dpi=300)
plt.plot(np.mean(f50, axis=0)[:100])
plt.xlabel("Iteration", size=14)
plt.ylabel("Accuracy (%)", size=14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.title("Training Curve for First 100 Iterations -\n Failure at 50", size=16)
plt.ylim(45, 85, 5)
plt.show()
plt.figure(dpi=300)
plt.plot(np.mean(f60, axis=0)[:100])
plt.xlabel("Iteration", size=14)
plt.ylabel("Accuracy (%)", size=14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.title("Training Curve for First 100 Iterations -\n Failure at 60", size=16)
plt.ylim(45, 85, 5)
plt.figure(dpi=300)
plt.show()
plt.figure(dpi=300)
# plt.plot(np.mean(f100, axis=0)[:110])
# plt.xlabel("Iteration", size=14)
# plt.ylabel("Accuracy (%)", size=14)
# plt.xticks(size=12)
# plt.yticks(size=12)
# plt.ylim(45, 85, 5)
# plt.title("Training Curve for First 110 Iterations -\n Failure at 100", size=16)
# plt.show()
```
